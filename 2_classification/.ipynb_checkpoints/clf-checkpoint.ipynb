# Naive Bayes (Наивный Байес)

Давайте сделаем простую классификацию писем на "спам/не спам". Например, на [этих данных](https://www.kaggle.com/uciml/sms-spam-collection-dataset)



````python
# ячейка импортов

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
from sklearn import feature_extraction, model_selection, naive_bayes, metrics, svm

from sklearn.neighbors import KNeighborsClassifier

# from IPython.display import Image
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline 
````

````python
! ls
````

````python
# читаем данные

data = pd.read_csv('spam.csv', encoding='latin-1') # обратите внимание на кодировку
data.head(10)
````

````python
# Почистим датафрейм

data = data.drop(['Unnamed: 2','Unnamed: 3','Unnamed: 4'],axis=1)

data = data.rename(columns={"v1":"label", "v2": "content"})


display(data)
# data.head(10)
````

давайте посчитаем распределение спама/не спама

````python
classes = pd.value_counts(data["label"])

print(classes)

classes.plot(kind = 'pie',  autopct='%0.f%%') 
# про string formatting https://realpython.com/python-string-formatting/

plt.title("Classes distribution")

plt.ylabel('')

plt.show()
````

**Анализ данных**

Давайте найдем частоты слов в сообщениях (отдельно для подкорпуса спама и не-спама). Для будущего алгоритма, слова будут признаками (features)

Для этого используем Counter из collections.

````python
# заберем все слова по условию, заджойним в строку, разобьем по пробелам, возьмем топ-20
count1 = Counter(" ".join(data[data['label']=='ham']["content"]).lower().split()).most_common(20) 
df1 = pd.DataFrame.from_dict(count1) # новый дф
df1 = df1.rename(columns={0: "words in non-spam", 1 : "count"})

# тот же набор операций для второго подкорпуса
count2 = Counter(" ".join(data[data['label']=='spam']["content"]).lower().split()).most_common(20)
df2 = pd.DataFrame.from_dict(count2)
df2 = df2.rename(columns={0: "words in spam", 1 : "count"})

display(df1,df2)
````

очень много "шума", давайте избавимся от стоп-слов. Используем [эту функцию](https://scikit-learn.org/stable/modules/generated/sklearn.feature_extraction.text.CountVectorizer.html) из sklearn 

````python
f = feature_extraction.text.CountVectorizer(stop_words = 'english')
type(f)
````

````python
data["content"] # посмотрим на тексты смс 
````

### как работает Count Vectorizer: пример на маленькой матрице

````python
from sklearn.feature_extraction.text import CountVectorizer

corpus = ['This is the first document.','Is this my the second document?']

vectorizer = CountVectorizer()
type(vectorizer)
X = vectorizer.fit_transform(corpus)
print(vectorizer.get_feature_names(),"\n")

display(X.toarray())

````

### идем дальше : создадим матрицу признаков X

````python
X = f.fit_transform(data["content"])
print(f.get_feature_names()[1000:1020]) # посмотрим на признаки,  например на 20 из них 

# display(X.shape)
display(X.toarray())
````

````python
display(X.shape) #  какого размера матрица X? 
````

метод ```.fit_transform()``` выучивает обучающие данные и возвращает матрицу формата "документ-термин" для всего датасета

Сейчас у нас 8404 признака (по числу уникальных слов в корпусе). Новый признак  j  в ряду i  равен 1 если слово j  появляется в тексте i . В ином случае значение признака = 0

**давайте попробуем предсказать:** окажется ли письмо спамом или не спамом

Следует помнить о двух нежелательных вариантах:

- спам-письмо классифицируется как нормальное (False Negative)
- обычное письмо классифицируется как спам (False positive)

Сначала, представим значения spam/ham как числовые (1,0) 

````python
data
````

````python
data["label"]=data["label"].map({'spam':1,'ham':0}) # в первой колонке переименуем значения 'spam' в 1  и 'ham' в 0

data
````

Затем разделим данные на обучающую и тестовую выборку [этой функцией](https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html)

Создадим новые переменные:
- X_train - кусочек матрицы "документ/термин", который мы "отдадим" алгоритму для обучения
- X_test - кусочек матрицы "документ/термин", на котором будем проверять, как алгоритм сработал 
- y_train - кусочек датафрейма с текстами, который мы отдадим для обучения
- y_test -  кусочек датафрейма с текстами, на котором будем проверять, как алгоритм сработал 

````python
# новые переменные
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, data['label'], test_size=0.2, random_state=50)

print(X_train.shape, X_test.shape)
sum(y_train == 1)
````

````python
print(y_train.shape, y_test.shape) # размерности таргета (y)
````

алгоритмов наивного байеса несколько, мы возьмем [Multinomial NB](https://scikit-learn.org/stable/modules/generated/sklearn.naive_bayes.MultinomialNB.html)

[дополнительная информация про Naive Bayes](https://nlp.stanford.edu/IR-book/pdf/13bayes.pdf)

Один из параметров модели -- это параметр регуляризации α (alpha)*([что это такое?](https://medium.com/syncedreview/applying-multinomial-naive-bayes-to-nlp-problems-a-practical-explanation-4f5271768ebf))*
Так как алгоритм быстрый, попробуем несколько вариантов с разным параметром регуляризации α:  зададим несколько значений альфы функцией range

````python
list_alpha = np.arange(1/100000, 20, 0.11) 
#возьмем лист равномерно распределенных значений от 1/100000 до 20
#можно выбрать другие числа,эти были выбраны экспериментально
print(len(list_alpha))
````

Затем проведем эксперименты: у нас 182 варианта возможных значений alpha. Для каждого  значения обучим алгоритм MultinomialNB,а также проверим метрики после обучения: precision, recall, accuracy 

````python
score_train = np.zeros(len(list_alpha)) # accuracy on train set
score_test = np.zeros(len(list_alpha)) #accuracy on test set
recall_test = np.zeros(len(list_alpha)) # recall on test
precision_test= np.zeros(len(list_alpha)) # precision on test
count = 0 
````

````python
for alpha in list_alpha: # для каждого значения в списке альф
    bayes = naive_bayes.MultinomialNB(alpha=alpha) # вызываем модель
    bayes.fit(X_train, y_train) # тренируем модель
    
    score_train[count] = bayes.score(X_train, y_train) # итерируемся по индексу в листе, добавляем значения accuracy on train

    
    score_test[count]= bayes.score(X_test, y_test) # то же самое,но для accuracy на тестовой выборке
 
    # а теперь для precision и recall
    recall_test[count] = metrics.recall_score(y_test, bayes.predict(X_test))
    precision_test[count] = metrics.precision_score(y_test, bayes.predict(X_test))
    
    count = count + 1 # после одной итерации, делаем новый шаг,  чтобы пройтись по всем alpha
````

Готово! У нас проведено 182 эксперимента. Найдем из них 10 лучших показателей

Сначала превратим наши списки с метриками в датафрейм,  по нему удобнее ориентироваться

````python
matrix = np.matrix(np.c_[list_alpha, score_train, score_test, recall_test, precision_test])

models = pd.DataFrame(data = matrix,
                      columns = ['alpha',
                                 'Train Accuracy',
                                 'Test Accuracy',
                                 'Test Recall',
                                 'Test Precision'])


display(models.head(10))

display(matrix.shape)
````


а теперь давайте оставим модель с наивысшей Accuracy on test

````python
best_index = models['Test Accuracy'].idxmax()
best_index

display(models.iloc[best_index])
````

Посмотрим, есть ли модели со 100% Precision

````python
models[models['Test Recall']>0.9].sort_values(by="Test Precision", ascending = False)
````

Давайте из них выберем ту, у которой одновременно precision выше 90% и наибольшее accuracy

````python
best_index = models[models['Test Precision']>0.9]['Test Accuracy'].idxmax()
best_index

print("best model's alpha is",models.iloc[best_index].alpha) # alpha-параметр у этой модели

models.iloc[best_index] # метрики лучшей модели
````

Так мы выбрали наилучшую модель из 182 проведенных экспериментов. Теперь давайте посмотрим на нее внимательнее: сколько раз она ошибается и где?

### попросим модель предсказать спам/не спам на тестовой выборке (предикшн)

````python
bayes = naive_bayes.MultinomialNB(alpha=0.330010, ) # используем лучшее значение alpha
bayes.fit(X_train, y_train) # тренируем модель

y_pred = bayes.predict(X_test) # предсказания модели здесь

#сделаем датафрейм с колонками для настоящего класса(это y_test) и для предсказанного (y_pred)

new_df = pd.DataFrame(data = y_test)
new_df["predicted"]=bayes.predict(X_test) # добавляем новую колонку с предсказаниями

new_df
````

давайте построим матрицу ошибок ([как работает матрица ошибок](https://en.wikipedia.org/wiki/Confusion_matrix))

````python
m_confusion_test = metrics.confusion_matrix(y_test, y_pred)


pd.DataFrame(data = m_confusion_test, columns = ['Predicted ham', 'Predicted spam'],
            index = ['Actual ham', 'Actual spam'])


````

````python
bayes.score(X_test, y_test) #accuracy on testset
````

Заметим, что так как мы определили лейбл spam как класс 1, наша табличка "перевернута" и начинается с True Negative (Actual ham x Predicted ham)

### Помимо Наивного Байеса: Support Vector Machines (SVM, метод опорных векторов)

попробуем сделать ту же задачу, с алгоритмом SVM.

У моделей SVM свой параметр регуляризации: вместо alpha появляется C ([что это?](https://www.analyticsvidhya.com/blog/2021/04/insight-into-svm-support-vector-machine-along-with-code/#:~:text=Soft%20margin%20SVM%20is%20implemented,has%20small%20values%20of%20C.https://www.analyticsvidhya.com/blog/2021/04/insight-into-svm-support-vector-machine-along-with-code/#:~:text=Soft%20margin%20SVM%20is%20implemented,has%20small%20values%20of%20C.))

````python
list_C = np.arange(500, 2000, 100) #100000

score_train = np.zeros(len(list_C))
score_test = np.zeros(len(list_C))
recall_test = np.zeros(len(list_C))
precision_test= np.zeros(len(list_C))
count = 0
for C in list_C:
    svc = svm.SVC(C=C)
    svc.fit(X_train, y_train)
    score_train[count] = svc.score(X_train, y_train)
    score_test[count]= svc.score(X_test, y_test)
    recall_test[count] = metrics.recall_score(y_test, svc.predict(X_test))
    precision_test[count] = metrics.precision_score(y_test, svc.predict(X_test))
    count = count + 1 
````

посмотрим на 10 первых моделей и их метрики:

````python
matrix = np.matrix(np.c_[list_C, score_train, score_test, recall_test, precision_test])
models = pd.DataFrame(data = matrix, columns = 
             ['C', 'Train Accuracy', 'Test Accuracy', 'Test Recall', 'Test Precision'])
models.head(n=10)
````

выберем модель с лучшим test precision

````python
best_index = models['Test Precision'].idxmax()

display(models.iloc[best_index, :])
````

````python
# посмотрим на модели с почти 100% Precision

models[models['Test Precision']>0.99].head(5)
````

````python
best_index = models[models['Test Precision']>0.99]['Test Accuracy'].idxmax()
svc = svm.SVC(C=list_C[best_index])
svc.fit(X_train, y_train)
models.iloc[best_index, :]
````

построим матрицу ошибок для SVM

````python
m_confusion_test = metrics.confusion_matrix(y_test, svc.predict(X_test))
pd.DataFrame(data = m_confusion_test, columns = ['Predicted 0', 'Predicted 1'],
            index = ['Actual 0', 'Actual 1'])
````
